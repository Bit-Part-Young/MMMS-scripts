```python
"""Buckingham-Coulomb potential实现"""

import numpy as np


def Buckingham_Coulomb(
    r: float, A: float, rho: float, C: float, q_1: float, q_2: float
) -> float:
    V = A * np.exp(-r / rho) - C / r**6 + q_1 * q_2 / r
    return V
```

```python
import matplotlib.pyplot as plt

# 参数待确定
A = 4.0
rho = 1.0
C = 1.0
# A = 18003.757
# rho = 1 / 4.87318
# C = 133.5381
q_1 = 1.0
q_2 = 1.0

distance_list = np.linspace(0.5, 5.0, 1000)
V_list = Buckingham_Coulomb(distance_list, A, rho, C, q_1, q_2)


plt.plot(distance_list, V_list, label="Buckingham potential", color="blue")
plt.xlabel("distance")
plt.ylabel("potential energy")
plt.grid(True)
plt.legend()
plt.show()
```
