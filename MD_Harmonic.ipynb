```python
import numpy as np

# reference code: Lingti Kong's code

"""一端固定的弹簧简谐振子的velocity verlet算法实现"""

def thermo_data(x, v, k=1.0, mass=1.0, step=0, dt=0.05):
    Ep = 0.5 * k * x**2
    Ek = 0.5 * mass * v**2
    Et = Ep + Ek

    return [step, step * dt, x, v, Ep, Ek, Et]


def MD_Harmonic():
    mass = 1.0
    k = 1.0
    x = 0.0
    f = -k * x
    v = 0.5
    a = f / mass

    period = 2.0 * np.pi * np.sqrt(mass / k)
    dt = 0.05
    half_dt = dt * 0.5
    nsteps = int(2.0 * period / dt)
    ioutput = 5

    thermo = []
    thermo_data_init = thermo_data(x, v)
    thermo.append(thermo_data_init)

    # velocity verlet算法实现
    for istep in range(1, nsteps + 1):
        v = v + a * half_dt
        x = x + v * dt
        f = -k * x
        a = f / mass
        v = v + a * half_dt

        # 每5步输出热力学信息
        if istep % ioutput == 0:
            thermo_data_step = thermo_data(x, v, step=istep)
            thermo.append(thermo_data_step)

    return thermo
```

```python
# 不用科学计数法显示
np.set_printoptions(suppress=True)
thermo = np.array(MD_Harmonic()).round(5)
thermo
```

```python
import matplotlib.pyplot as plt

plt.xlabel("MD time")
plt.ylabel("distance")
plt.plot(thermo[:, 1], thermo[:, 2], marker="o", label="x")

plt.legend()

plt.show()
```
