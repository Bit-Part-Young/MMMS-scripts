```python
import matplotlib.pyplot as plt
import pandas as pd

thermo_data = pd.read_csv('thermo_data.csv')

plt.xlabel("MD time")
plt.ylabel("Energy")
plt.plot(thermo_data.iloc[:, -1], thermo_data.iloc[:, 1], marker="o", label="$E_p$")
plt.plot(thermo_data.iloc[:, -1], thermo_data.iloc[:, 0], marker="o", label="$E_k$")
plt.plot(thermo_data.iloc[:, -1], thermo_data.iloc[:, 2], marker="o", label="$E_t$")

plt.legend()

plt.show()
```
